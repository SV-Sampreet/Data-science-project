```python
plt.figure(figsize=(20,18))
sns.pairplot(movie_data)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15, 12))
sns.scatterplot(x='Name', y='Rating', data=movie_data.head(90), color='g')
plt.xticks(rotation=90)
plt.title('Ratings of movies')
plt.xlabel('Movie Name')
plt.ylabel('Ratings')
plt.show()
```

```python
plt.figure(figsize=(14, 12))
sns.lineplot(x='Director', y='Votes', data=movie_data.head(100))
plt.xticks(rotation=90)
plt.title('Votes vs Director')
plt.xlabel('Director Names')
plt.ylabel('Votes')
plt.show()
```

```python
movie_data.columns
```

```python
genre_counts = movie_data['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

top_n_genres = genre_counts.head(5)
top_n_genres
```

```python
plt.figure(figsize=(4, 4))
plt.pie(top_n_genres['Count'], labels=top_n_genres['Genre'], autopct='%1.1f%%', 
        startangle=140, colors=sns.color_palette('pastel'))
plt.title('Movie Genre Distribution', fontsize=16)
plt.axis('equal')
plt.show()
```

```python
director_avg_rating = movie_data.groupby('Director')['Rating'].mean().reset_index()

director_avg_rating = director_avg_rating.sort_values(by='Rating', ascending=False)

top_directors = director_avg_rating.head(10)
top_directors
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_directors, x='Rating', y='Director', palette='viridis', hue='Director', legend=False)

plt.title('Top Directors by Average Rating', fontsize=16)
plt.xlabel('Average Rating', fontsize=14)
plt.ylabel('Director', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=movie_data, x='Rating', y='Votes')

plt.title('Votes vs. Rating', fontsize=16)
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Votes', fontsize=14)

plt.show()
```

```python
plt.figure(figsize=(15, 12))
sns.lineplot(data=movie_data.head(100), x='Actor 1', y='Rating', color='blue')
sns.lineplot(data=movie_data.head(100), x='Actor 2', y='Rating', color='black')
sns.lineplot(data=movie_data.head(100), x='Actor 3', y='Rating', color='red')
plt.legend(title='Legend', labels=['Actor 1', 'Actor 2', 'Actor 3'])
plt.xticks(rotation=90)
plt.title('Ratings vs Actors')
plt.xlabel('Actors')
plt.ylabel('Ratings')
plt.show()
```

```python
plt.figure(figsize=(15, 12))
sns.barplot(data=movie_data.head(200), x='Genre', y='Rating')
plt.xticks(rotation=90)
plt.title('Ratings vs Genre')
plt.xlabel('Genre')
plt.ylabel('Ratings')
plt.show()
```
